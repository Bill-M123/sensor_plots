```python
import pandas as pd
import numpy as np
import plotly
import matplotlib.pyplot as plt
import random

import scipy.fftpack
```

## Create sample data

start with a square wave, 1 reading per second, amplitude 1, 50% duty cyle, 1 hour of data

```python
## Create sample data

period=100
n_periods=1


x=[x for x in range(3600)]
y=[]
for i in range(36):
    for j in range(int(period/2)):
        if i%2 ==0:
            y.append(0)
        else:
            y.append(1)

x=x[:period*n_periods]
y=y[:period*n_periods]

def plot_first_cycle(x,y):
    plt.figure(figsize=(10,3),dpi=100)
    plt.scatter(x,y)
    plt.xlim(0,n_periods*period)
    plt.title('First cycle')
    plt.show()
    return

plot_first_cycle(x,y)
```

## What does the fft look like for a normal cycle?

```python
def show_fft(x,y):
    x_test=x[:int(len(x)/2)]
    yf = scipy.fftpack.fft(y)
    power=np.abs(yf[:int(len(x)/2)])

    plt.figure(figsize=(10,3),dpi=100)
    plt.plot(x_test, power)

    plt.show()
    return power

no_hammer=show_fft(x,y)
```

##  Let's simulate a hammer drop.  Assume 50% magnitude that gets captured for 1 second at time 50.

```python
y_hammer=y.copy()

stepsize=0.35
y_hammer[24]=stepsize
y_hammer[25]=stepsize
y_hammer[26]=stepsize
y_hammer[27]=stepsize
plot_first_cycle(x,y_hammer)
```

```python
w_hammer=show_fft(x,y_hammer)
plt.plot(x[:int(len(x)/2)],(no_hammer-w_hammer))
plt.show()
```

## Add noise

```python
random.seed(42)
random.random()

range_max=0.2
range_min=0-range_max

y_noise=[x+random.uniform(range_min,range_max) for x in y]
y_noise2=[x+random.uniform(range_min,range_max) for x in y]
y_noise3=[x+random.uniform(range_min,range_max) for x in y]
y_hammer_noise=[x+random.uniform(range_min,range_max) for x in y_hammer]

plot_first_cycle(x,y_noise)
plot_first_cycle(x,y_noise2)
plot_first_cycle(x,y_noise3)
plot_first_cycle(x,y_hammer_noise)
```

```python
fft_noise_power=show_fft(x,y_noise)
fft_noise_power2=show_fft(x,y_noise2)
fft_noise_power3=show_fft(x,y_noise3)
fft_hammer_noise_power=show_fft(x,y_hammer_noise)

def plot_difference(x,y,title=''):

    plt.figure(figsize=(10,3),dpi=100)
    plt.plot(x,y)
    plt.title(title)
    plt.show()
    
def plot_comp(x1,y1,x2,y2,label1='lable1',label2='lable2',color1='blue',color2='red',title='Title'):
    plt.figure(figsize=(10,3),dpi=100)
    plt.plot(x1,y1,label=label1,color=color1)
    plt.plot(x2,y2,label=label2,color=color2)
    plt.title(title)
    plt.legend(loc='best')
    plt.show()
    
plot_x=x[:int(len(x)/2)]

plot_difference(plot_x,(fft_noise_power-fft_noise_power2),title='Baseline')
plot_difference(plot_x,(fft_noise_power-fft_hammer_noise_power),title='Hammer_drop')
plot_comp(plot_x,(fft_noise_power-fft_noise_power2),plot_x,(fft_noise_power2-fft_noise_power3),
          label1='Noise1/2',label2='Noise2/3',title='Comparison: 3 Noisey Signals')
plot_comp(plot_x,(fft_noise_power2-fft_noise_power3),plot_x,(fft_noise_power-fft_hammer_noise_power),
          label1='Noise1',label2='Hammerdrop',title='Comparison: 2 noisy signals vs 1 noisy signal and a hammer drop')
plot_comp(plot_x,np.abs((fft_noise_power2-fft_noise_power3)),plot_x,np.abs((fft_noise_power2-fft_hammer_noise_power)),
          label1='Baseline',label2='Hammerdrop',title='Comparison: 2 noisy signals vs 1 noisy signal and a hammer drop')
```
